```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array, load_img

# Create the main application window
my_w = tk.Tk()
my_w.geometry("600x500")
my_w.title('Wildlife Conservation Using Image Recognition')

# Load the background image
bgimg = ImageTk.PhotoImage(file="C:/Users/deves/Desktop/Training/First poject/1.jpg")

# Create a label to display the background image
bg_label = tk.Label(my_w, image=bgimg)
bg_label.place(relwidth=1, relheight=1)

# Create a label with your title
my_font1 = ('times', 18, 'bold')
title_label = tk.Label(my_w, text='Wildlife Conservation Using Image Recognition', font=my_font1)
title_label.pack(pady=20)

# Load your trained model
model = tf.keras.models.load_model('wildlife_model.h5')

# Define a mapping of class indices to class names
class_names = {
    0: 'Class 0',
    1: 'Class 1',
    # Add more class mappings as needed
}

# Create a function to upload, classify, and display an image
def classify_image():
    f_types = [('Image Files', '*.jpg *.jpeg *.png *.bmp')]
    filename = filedialog.askopenfilename(filetypes=f_types)
    if filename:
        image = load_img(filename, target_size=(224, 224))
        image = img_to_array(image)
        image = np.expand_dims(image, axis=0)
        image = image / 255.0  # Normalize the image

        prediction = model.predict(image)
        class_index = np.argmax(prediction)
        class_name = class_names.get(class_index, 'Unknown')

        result_label.config(text=f"Predicted Class: {class_name}")

        # Display the uploaded image
        uploaded_img = Image.open(filename)
        uploaded_img = uploaded_img.resize((300, 300), Image.ANTIALIAS)
        uploaded_img = ImageTk.PhotoImage(uploaded_img)
        uploaded_label = tk.Label(my_w, image=uploaded_img)
        uploaded_label.image = uploaded_img
        uploaded_label.pack(pady=20)

# Create a button to upload an image and classify
upload_button = tk.Button(my_w, text='Upload and Classify Image', width=30, command=classify_image)
upload_button.pack()

# Create a label to display the classification result
result_label = tk.Label(my_w, text="", font=("Helvetica", 16))
result_label.pack(pady=10)

my_w.mainloop()  # Keep the window open
```
